```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Encryption', 'Data Mgmt', 'Auth/Session', 'Transparency', 'Robustness']
x = np.arange(len(categories))
width = 0.2

cg_tech = [83.5, 90.0, 75.0, 83.5, 80.0]
gm_tech = [66.5, 80.0, 75.0, 100.0, 40.0]
cg_subj = [41.75, 30.75, 37.5, 54.15, 45.75]
gm_subj = [38.25, 31.75, 40.75, 52.0, 50.0]

fig, ax = plt.subplots(figsize=(10, 5))

rects1 = ax.bar(x - 1.5*width, cg_tech, width, label='ChatGPT Tech', color='#16A085')
rects2 = ax.bar(x - 0.5*width, gm_tech, width, label='Gemini Tech', color='#4285F4')
rects3 = ax.bar(x + 0.5*width, cg_subj, width, label='ChatGPT User', color='#95A5A6')
rects4 = ax.bar(x + 1.5*width, gm_subj, width, label='Gemini User', color='#BDC3C7')

ax.set_ylabel('Security Index', fontsize=12, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=11, fontweight='bold')
ax.set_ylim(0, 130)

ax.legend(ncol=4, loc='upper center', bbox_to_anchor=(0.5, 1.15), fontsize=10, frameon=False)

def autolabel(rects, color):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=10, fontweight='black', color=color, rotation=90) 

autolabel(rects1, '#16A085')
autolabel(rects2, '#4285F4')
autolabel(rects3, '#7F8C8D')
autolabel(rects4, '#95A5A6')

ax.yaxis.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
